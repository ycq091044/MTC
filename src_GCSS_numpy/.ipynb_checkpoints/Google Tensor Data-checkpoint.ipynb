### load data

```python
import pandas as pd

data = pd.read_csv('/mnt/disks/ssd/tensorData/google-search-trends.csv')
data
```

### index of three modes

```python
# all zipcode
zipcode = list(filter(lambda x: ('US' in x) and (x.count('_') == 2), data.key.unique()))

# the time
time = data.date.unique().tolist()

# disease list
disease = data.columns.tolist()[2:]
```

```python
len(zipcode), len(time), len(disease)
```

### extract the tensor

```python
import numpy as np

# tensor placeholder
T = np.zeros((len(zipcode), len(disease), len(time)))

# process per zipcode
for i, single_zip in enumerate(zipcode):
    print (i, single_zip)
    tmp = data[data.key == single_zip]
    re_time = [time.index(i) for i in tmp.date.tolist()]
    T[i, :, :][:, re_time] = tmp.values[:, 2:].T
```

```python
# change nan to zero
T = np.nan_to_num(T)
```

### aggregation matrix

```python
import pickle

# zipcode -> state mapping
state = np.unique([item.split('_')[1] for item in zipcode]).tolist()
P1 = np.zeros((len(state), len(zipcode)))

for i, z in enumerate(zipcode):
    for j, s in enumerate(state):
        if s in z:
            P1[j, i] = 1
            
# date -> week mapping
P3 = np.zeros(((len(time) - 1) // 7 + 1, len(time)))
for i, t in enumerate(time):
    P3[i // 7, i] = 1
    
```

### coarse tensor

```python
O1 = np.einsum('ijk,ri->rjk',T,P1,optimize=True)
O3 = np.einsum('ijk,rk->ijr',T,P3,optimize=True)
```

```python
print (O1.shape, O3.shape, T.shape)

pickle.dump(O1, open('/mnt/disks/ssd/tensorData/data/release/O1_google.pkl', 'wb'))
pickle.dump(O3, open('/mnt/disks/ssd/tensorData/data/release/O3_google.pkl', 'wb'))
pickle.dump(T, open('/mnt/disks/ssd/tensorData/data/release/T_google.pkl', 'wb'))
pickle.dump(P1, open('/mnt/disks/ssd/tensorData/data/release/P1_google.pkl', 'wb'))
pickle.dump(P3, open('/mnt/disks/ssd/tensorData/data/release/P3_google.pkl', 'wb'))
```
